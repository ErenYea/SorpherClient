```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.support.ui import Select

from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from google.oauth2 import service_account


driver.get("https://racing.hkjc.com/racing/information/English/Racing/DisplaySectionalTime.aspx?RaceDate=06/03/2022&RaceNo=1")
driver.implicitly_wait(30)
driver.maximize_window()

options = driver.find_elements_by_xpath("//span[@class='f_fr']/select/option")
options.pop(0)

data=[]

for i,opt in enumerate(options):
    driver.refresh()
    value = driver.find_element_by_xpath(f"//span[@class='f_fr']/select/option[{i+2}]").text
    select = Select(driver.find_element_by_xpath("//select[@id='selectId']"))
    select.select_by_visible_text(value)
    click = driver.find_element_by_xpath("//a[@id='submitId']").click()
    
    print(value)

# print(len(options))

    pages = driver.find_elements_by_xpath("//div[@class='commTitlePic f_clear']/table/tbody/tr/td[2]/table/tbody/tr/td")
    pages.pop(0)
    print(len(pages))

    for ind,l in enumerate(pages):
        try:
            driver.refresh()
            if (ind!=0):

                link = driver.find_element_by_xpath(f"//div[@class='commTitlePic f_clear']/table/tbody/tr/td[2]/table/tbody/tr/td[{ind+1}]/a").get_attribute('href')
                driver.get(link)
                driver.implicitly_wait(10)
                print("***********************************")

            else:
                print("------------------------------------------------------------------------------------------------------")
        
        
            data1 = driver.find_element_by_xpath("//div[@class='search']/p/span").text.split(" ")
            date = data1[1].split(':')[1].split(',')[0]
            course_name = data1[2]+ ' ' + data1[3]
            race_number = driver.find_element_by_xpath("//div[@class='Race f_clear']/p").text
            data2 = driver.find_element_by_xpath("//div[@class='Race f_clear']/p[2]").text.split('-')
            print(data2)
            classs = data2[0]
            distance = data2[1]
            going = data2[-1].split('\n')[0]
            finishing_order = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td").text
            horse = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[3]").text
            first_sec = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[4]/p[2]").text
            second_sec = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[5]/p[2]").text
            third_sec = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[6]/p[2]").text
            try:
                fourth_sec = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[7]/p[2]").text
            except:
                fourth_sec = "NIL"
            try:
                fifth_sec = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[8]/p[2]").text
            except:
                fifth_sec = "NIL"
            try:
                sixth_sec = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[9]/p[2]").text
            except:
                sixth_sec = "NIL"
            time = driver.find_element_by_xpath("//table[@class='table_bd f_tac race_table']/tbody/tr/td[10]").text

            minutes = int(time.split(":")[0])
            add = float(time.split(":")[1])
            conv_time = round((minutes*60+add),2)

            data.append([date,course_name,race_number,classs,distance,going,finishing_order,horse,first_sec,second_sec,third_sec,fourth_sec,fifth_sec,sixth_sec,conv_time])
            
        except:
            
            print("Information will be released shortly")
            
SERVICE_ACCOUNT_FILE = 'keys.json'
SCOPES = ['https://www.googleapis.com/auth/spreadsheets']

creds = None
creds = service_account.Credentials.from_service_account_file(
        SERVICE_ACCOUNT_FILE, scopes=SCOPES)

SAMPLE_SPREADSHEET_ID = '19gNgMJPeriXMJ2mRnBrulMO1apYnBBbBBGHZMhwVF2M'


service = build('sheets', 'v4', credentials=creds)

# Call the Sheets API
sheet = service.spreadsheets()

# Write
request = sheet.values().update(spreadsheetId=SAMPLE_SPREADSHEET_ID,
                            range="Sheet2!A2", valueInputOption="USER_ENTERED", body={"values":data}).execute()
```
