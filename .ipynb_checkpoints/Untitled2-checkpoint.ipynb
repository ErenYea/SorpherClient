```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.support.ui import Select

a = ActionChains(driver)

driver.get("https://www.attheraces.com/results")
driver.implicitly_wait(10)
driver.maximize_window()

interested_courses = ['Ascot','Bangor-on-dee','Bath','Brighton','Chepstow','Chester','Doncaster','Fakenham','Ffos las','Fontwell','Hereford','Hexham','Lingfield','Newcastle','Newton abbot','Plumpton','Ripon','Sedgefield','Southwell','Uttoxeter','Windsor','Wolverhampton','Worcester','Yarmouth']


time.sleep(3)
data = []
for day in range(15):
    
    if (day==0):
        previous = driver.find_element_by_xpath("//button[@class='button button--small datepicker__control datepicker__control--prev']")
        previous.click()

        time.sleep(2)
        
    else:
        race_courses = driver.find_elements_by_xpath("//div[@id='tab-race-nav--uk']/div/div/div/div/div")

        for i,race_course in enumerate(race_courses):
            

                course_name = driver.find_element_by_xpath(f"//div[@id='tab-race-nav--uk']/div/div/div/div/div[{i+1}]/label/span").text

                if course_name.strip() in interested_courses:

                    races = driver.find_elements_by_xpath(f"//div[@id='tab-race-nav--uk']/div/div/div/div/div[{i+1}]/div/div/ul/li")
                    

                    for j,rac in enumerate(races):
                        race = rac.find_element_by_tag_name('a').get_attribute('href')
                        print(race)
                        data.append(race)

        driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.HOME)
        
        date_input = driver.find_element_by_xpath("//input[@id='race-nav-datepicker']")
        a.move_to_element(date_input).perform()
        
        previous = driver.find_element_by_xpath("//button[@class='button button--small datepicker__control datepicker__control--prev']")
        previous.click()

        time.sleep(2)
        

print(data)
```

```python
col = 0
for i in all_columns:
    if all_columns.index(i) <=2 :
        continue
    if i.text == '':
        ele =  driver.find_element_by_xpath('//div[@class="js-scrollbar__track scrollbar__track"]')
        desired_y = (ele.size['height'] / 2) +ele.location['y']
        window_h = driver.execute_script('return window.innerHeight')
        window_y = driver.execute_script('return window.pageYOffset')
        current_y = (window_h / 2) + window_y
        scroll_y_by = desired_y - current_y
        driver.execute_script("window.scrollBy(0, arguments[0]);", scroll_y_by)
        horizontal_bar_width = driver.find_element_by_xpath('//div[@class="js-scrollbar__track scrollbar__track"]').rect['width']
        slider = driver.find_element_by_xpath('//div[@class="js-scrollbar__track scrollbar__track"]')
        ActionChains(driver).click_and_hold(slider).move_by_offset(horizontal_bar_width/2, 0).release().perform()
        
    ele =  i
    desired_y = (ele.size['height'] / 2) +ele.location['y']
    window_h = driver.execute_script('return window.innerHeight')
    window_y = driver.execute_script('return window.pageYOffset')
    current_y = (window_h / 2) + window_y
    scroll_y_by = desired_y - current_y
    driver.execute_script("window.scrollBy(0, arguments[0]);", scroll_y_by)
    if i.text == 'Finish':
        print('ok')
    print(i.text)
    col += 1
```

```python

race_courses = driver.find_elements_by_xpath("//div[@id='tab-race-nav--uk']/div/div/div/div/div")

for i,race_course in enumerate(race_courses):
    if (i>=1):
        
        course_name = driver.find_element_by_xpath(f"//div[@id='tab-race-nav--uk']/div/div/div/div/div[{i+1}]").text
        
        if course_name.strip() in interested_courses:
        
            races = driver.find_elements_by_xpath(f"//div[@id='tab-race-nav--uk']/div/div/div/div/div[{i+1}]/div/div/ul/li")
            data = []
            
            for j,rac in enumerate(races):
                try:
                    race = driver.find_element_by_xpath(f"//div[@id='tab-race-nav--uk']/div/div/div/div/div[{i+1}]/div/div/ul/li[{j+1}]/a").get_attribute('href')
                    print(race)
                    driver.get(race)
                    sectional_times = driver.find_element_by_xpath("//ul[@class='tabs__links tabs__links--grey js-tabs__links']/li[2]/a").get_attribute('href')
                    driver.get(sectional_times)
                    driver.refresh()

                    total_runners = driver.find_element_by_xpath("//div[@class='race-header__details race-header__details--primary']/p[3]").text.split('-')[1].split(' ')[1]

                    for k in range(int(total_runners)):
                        horse_name = driver.find_element_by_xpath(f"//div[@class='card-body']/div[{k+1}]/div/div[3]/div/div/h2/a")   
                        print(horse_name.text)
                    print('Sectional Times')
                    driver.back()
                except:
                    continue
            print(len(races))
            print(course_name)
```
